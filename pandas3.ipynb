```python
weather.shape    # wide data, or rectangular data
```

```python
weather_melt = pd.melt(weather, id_vars=["year", "month", "element"], var_name="day", value_name="temp")
```

```python
weather_melt
```

```python
weather_tidy = weather_melt.pivot_table(index=["year", "month", "day"], columns="element", values="temp" )
```

```python
# melt --> melt the columns into a single column, columns ---> column values in rows
# pivot_table --> converts values of a column into seperate columns
```

```python
weather_tidy
```
